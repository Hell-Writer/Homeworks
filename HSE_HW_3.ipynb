В таблице table.tsv лежат посылки эконома в контест, сделанные в течение второй самостоятельной работы. В табличке три колонки разделенные знаком табуляции:

- ID - представляет из себя номер посылки

- user - анонимизированный id студента

- status - была ли сдана задача (OK/TL/WA/RE и тп)

Посчитайте для каждой задачи как часто на ней возникала ошибка WA. Сохраните ответ в файл output.tsv в формате таблицы, где колонки разделены друг от друга знаком табуляции.

```python
import csv
status=[]
task=[]
d={}
with open('table.tsv','r') as f:
    l=f.readline()
    l=l.strip()
    indx=l.split('\t').index('status')
    while True:
        s=f.readline()
        if s=='':
            break
        row=s.split('\t')[indx]
        status.append(row)
        
with open('table.tsv','r') as f:
    l=f.readline()
    
    l=l.strip()

    indx=l.split('\t').index('task')
    while True:
        s=f.readline()
        if s=='':
            break
        row=s.split('\t')[indx]
        task.append(row.strip())

for item, jtem in zip(status,task):
    d.setdefault(jtem,0)
    if item=='WA':
        
        d[jtem]+=1

d=dict(sorted(d.items()))
with open('output.tsv','w') as f:
    t=csv.writer(f, delimiter='\t')
    t.writerow(['task', 'status_cnt'])
    for ktem in d:
        t.writerow([ktem, d[ktem]])
```

По результатам полугодия команда Пандаса Фигандаса не смогла выполнить kpi, который он им поставил. Он задумался, а нет ли в его команде бездельников? Он смог найти файл с последними событиями в репозитории и собирается посчитать научно обоснованный показатель эффективности разработчика - количество написанного кода (в строках), количество добавленных файлов и количество комитов. Но вот же беда - данные ведь в формате json! Помогите ему и предоставьте таблицу в формате .tsv! Таблица должна содержать колонки в следующем порядке

["username", "commits", "changed_lines", "new_files"],

а также должна быть отсортирована по username.

- username - юзернейм разработчика

- commits - количество комитов разработчика

- changed_lines - суммарное количество changed_lines

- new_files - суммарное количество новых файлов добавленных разработчиком. Новым файлом считается любой файл, которого вы раньше не встречали (если смотреть на commit_time). Обратите внимание, что данные изначально не отсортированы по commit_time

```python
users={}
files={}
import json
import csv
from collections import Counter
with open('input.json') as js:
    data=json.load(js)
for item in data:
    username=item['username']
    users.setdefault(username,[0,0,0])
    users[username][0]+=1
    a=[item['files'][i]['changed_lines'] for i in range(len(item['files'])) ]
    users[username][1]+=sum(a)
    comtim=item['commit_time']
    for ktem in item['files']:
        files.setdefault(ktem['name'],(username,comtim))
        if comtim<=files[ktem['name']][1]:
            files[ktem['name']]= (username,comtim)
uniq=[i[0] for i in files.values()]
cnt=Counter(uniq)
for item in cnt:
    users[item][2]=cnt[item]
users=list(users.items())
users.sort()
userlist=[item[0] for item in users]
comlist=[item[1][0] for item in users]
changelist=[item[1][1] for item in users]
newfilelist=[item[1][2] for item in users]
table=list(zip(userlist,comlist,changelist,newfilelist))

with open('output.tsv','w') as f:
    t=csv.writer(f, delimiter='\t')
    t.writerow(["username", "commits", "changed_lines", "new_files"])
    for ktem in table:
        t.writerow(ktem)
```
