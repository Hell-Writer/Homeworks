```python
#Каждый из N школьников некоторой школы знает M языков. 
#Определите, какие языки знают все школьники и языки, которые знает хотя бы один из школьников.
from collections import Counter
a=int(input())
i, cnt=0, 0
d, f, l=[],[],[]
while i<a:
    b=int(input())
    j=0
    while j<b:
        c=input()
        d.append(c)
        j+=1
    i+=1
g=set(d)
cnt=Counter(d)
l=[item for item,jtem in cnt.items() if int(jtem)==a ]
print(len(l))
if len(l)>0:
    for lang in l:
        print(lang)
print(len(g))
for language in g:
    print(language)
```

```python
#В олимпиаде по информатике принимало участие несколько человек. 
#Победителем олимпиады становится человек, набравший больше всех баллов. 
#Победители определяются независимо по каждому классу. 
#Определите количество баллов, которое набрал победитель в каждом классе. 
with open('input.txt',encoding='UTF-8') as f:
    lines = f.readlines()
S9, S10, S11 = [], [], []
lines = [line.rstrip() for line in lines]
for item in lines:
    a=item.split(sep=' ')
    if a[2]=='9':
        S9.append(int(a[3]))
    elif a[2]=='10':
        S10.append(int(a[3]))
    else: 
        S11.append(int(a[3]))
b=max(S9)
c=max(S10)
d=max(S11)
print(b, c, d)
```

```python
#MVP для приложения "Пенькофф Инвестиции"
r=int(input())
i=0
l=[]
sumlist=[]
clients={}
companies={}
while i<r:
    l.append(input().split(sep=' '))
    i+=1
for item in l:
    if item[0]=='BUY':
        companies.setdefault(item[2], 1)
        clients.setdefault(item[1], dict())
        clients[item[1]].setdefault(item[2], 0)
        clients[item[1]][item[2]]+=int(item[3])
    if item[0]=='SELL':
        companies.setdefault(item[2], 1)
        clients.setdefault(item[1], dict())
        clients[item[1]].setdefault(item[2], 0)
        clients[item[1]][item[2]]-=int(item[3])
        if clients[item[1]][item[2]]<0:
            clients[item[1]][item[2]]=0
    if item[0]=='PRICE_RAISE':
        companies.setdefault(item[1], 1)
        companies[item[1]]*=(1+float(item[2])/100)
    if item[0]=='PRICE_FALL':
        companies.setdefault(item[1], 1)
        companies[item[1]]*=(1-float(item[2])/100)
    if item[0]=='BALANCE':
        clients.setdefault(item[1], dict(ErrorTimeParadox=0))
        companies.setdefault('ErrorTimeParadox', 1)
        s=[jtem for jtem in clients[item[1]]]
        for ktem in s:
            a=(companies[ktem])*int(clients[item[1]][ktem])
            sumlist.append(a)
        print(round(sum(sumlist)))
        sumlist=[]
```

```python
#Сжатие последовательности
a=input()+' -1'
a=list(a.split(sep=' '))
i=0
l=[]
m=[]
while i<len(a)-1:
    if int(a[i])+1==int(a[i+1]):
        l.append(a[i])
        l.append(a[i+1])
    elif len(l)>2:
        b=str(l[0])+'-'+str(l[-1])
        m.append(b)
        l=[]
    elif len(l)>0:
        m.append(l[0])
        m.append(l[1])
        l=[]
    else:
        m.append(a[i])
    i+=1
print(','.join(m))
```

```python
#Выборы в Государственную Думу
i=0
with open('input.txt',encoding='UTF-8') as f:
    lines = f.readlines()
l=[x.strip() for x in lines]
pl=[x.rsplit(' ', 1) for x in l]
party_pos=[[x,int(y)] for [x,y] in pl]
votes=[y for [x,y] in party_pos]
fnum=sum(votes)/450
first_round={jtem:int((item/fnum)//1) for jtem, item in party_pos}
second_round=[(jtem, (item/fnum)%1) for jtem, item in party_pos]
second_round=sorted(second_round, reverse=True, key=lambda x: x[1])
dif=450-sum(first_round.values())
while i<dif:
    first_round[second_round[i][0]]+=1
    i+=1
for item in first_round:
    print(item, first_round.get(item))
```
